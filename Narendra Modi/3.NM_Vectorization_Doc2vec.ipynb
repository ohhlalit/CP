```python
import pandas as pd
import nltk 
from nltk.stem import WordNetLemmatizer
import gensim
from gensim.models import Doc2Vec
```

```python
#to covert float to a string type as it is the required input for 
df['Cleaned_Speech_string'] = df['Cleaned_Speech_string'].astype(str)
```

```python
#required input was list of list of words of a sentance
corpus_tokenized = [nltk.word_tokenize(sentence) for sentence in df["Cleaned_Speech_string"]]
```

```python
len(corpus_tokenized)
```

#### Vectorisation using Doc2Vec
Doc2Vec is an algorithm for generating vector representations (embeddings)  of documents in a high - dimensional space. This technique allows us to capture semantic similarities and relationships between the documents. 

```python
from gensim.models.doc2vec import Doc2Vec, TaggedDocument
tagged_docs = [TaggedDocument(words=doc, tags=[i]) for i, doc in enumerate(corpus_tokenized)]
```

```python
tagged_docs[0]
```

```python
len(tagged_docs)
```

```python
#TRAINING THE MODEL OVER OUR DATA
model = Doc2Vec(tagged_docs, vector_size=300, window=5, min_count=5, workers=4, epochs=20)
```

#### Numerical representation of Text


```python
vector = model.infer_vector(corpus_tokenized[0])
```

```python
corpus_tokenized[0]
```

```python
vector
```

```python
import pandas as pd

# Create an empty DataFrame with the desired column names
df = pd.DataFrame(columns=['index', 'vector'])

# Iterate over the corpus_tokenized list and generate a vector for each element
for i, speech in enumerate(corpus_tokenized):
    vector = model.infer_vector(speech)
    # Store the index number and vector in a dictionary
    data = {'index': i, 'vector': vector}
    # Add the dictionary as a new row in the DataFrame
    df.loc[len(df)] = data

```

```python
# Create new columns for each element of the vector
for i in range(300):
    col_name = f'dim_{i+1}'
    df[col_name] = 0.0

# Iterate over each row in the DataFrame and assign the ith element of the vector to the corresponding column
for i, row in df.iterrows():
    for j, element in enumerate(row['vector']):
        col_name = f'dim_{j+1}'
        df.at[i, col_name] = element
```
