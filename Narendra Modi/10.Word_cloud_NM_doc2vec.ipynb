```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
#Load necessary Libraries
import pandas as pd
import numpy as np
import pickle
from ipywidgets import interact, Dropdown
import random

```

```python
df_data = pd.read_csv(r'/content/word_cloud.csv')
```

```python
df_data['Topics'] = df_data['Topics'].astype(str)
```

```python
df_data
```

```python
import re

words_to_remove = ['is','s','need', 'role', 'png','benefit' ,'India','world','help','global','will' , 'Indian','potential', 'indian', 'government', 'country', 'india', 'people', 'development', 'subject', 'text', 'AKT', 'india', 'idea', 'always', 'a', 'A.', 'been', "Indian's", 'government','important', 'the', 'author', 'how', 'central', 'theme', 'of', 'exploring', 'importance', 'impact', 'can', 'be', 'for', 'and', 'used', 'that', 'have']

def clean_topics(text):
    if isinstance(text, str):
        cleaned_text = ' '.join([word for word in re.split(r'\W+', text) if word.lower() not in words_to_remove])
        return re.sub(r'\s+', ' ', cleaned_text).strip()
    return ''

df_data['CleanedTopics'] = df_data['Topics'].apply(clean_topics)
```

```python

# Define the function to generate the word cloud
def generate_word_cloud(cluster, tenure):
    cluster_data = df_data[(df_data['cluster'] == cluster) & (df_data['tenure'] == tenure)]
    text = ' '.join(cluster_data['CleanedTopics'])
    
    # Define a custom color function
    def random_color_func(word=None, font_size=None, position=None, orientation=None, font_path=None, random_state=None):
        h = int(360.0 * random.random())
        s = int(100.0 * random.random())
        l = int(50.0 + 10.0 * random.random())
        return "hsl({}, {}%, {}%)".format(h, s, l)
    
    # Generate the word cloud with custom colors and white background
    wordcloud = WordCloud(color_func=random_color_func, background_color='white').generate(text)
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.show()

# Define the unique clusters and tenures
clusters = df_data['cluster'].unique()
tenures = df_data['tenure'].unique()

# Create the interactive dropdown menu
@interact(cluster=Dropdown(options=clusters), tenure=Dropdown(options=tenures))
def generate_word_cloud_interactive(cluster, tenure):
    generate_word_cloud(cluster, tenure)
```
