```python
#Importing necessary libraries 
import pandas as pd
import nltk 
from nltk.stem import WordNetLemmatizer
import gensim
from gensim.models import Doc2Vec
import pickle
```

```python
# Loading the datafile 
df = pd.read_csv("D:\PM_MANMOHAN_Speech_Cleaned_string (1).csv")
```

```python
PM_MANMOHAN_Speech_Cleaned_string = open("PM_MANMOHAN_Speech_Cleaned_string.pickle", "wb")
pickle.dump(df, PM_MANMOHAN_Speech_Cleaned_string)
PM_MANMOHAN_Speech_Cleaned_string.close()
```

```python
df
```

```python
#to covert float to a string type as it is the required input for 
df['Cleaned_Speech_string'] = df['Cleaned_Speech_string'].astype(str)
```

```python
# using the NLTK library to tokenize the sentences 
corpus_tokenized = [nltk.word_tokenize(sentence) for sentence in df["Cleaned_Speech_string"]]
```

```python
len(corpus_tokenized)
#sanity check should be equal to no. of speeches
```

```python
corpus_tokenized[0]
```

```python
# The code snippet is using the gensim library to create tagged documents for Doc2Vec model training.
# The output of this code is a list of TaggedDocument objects called "tagged_docs" that can be used to train a Doc2Vec model.
from gensim.models.doc2vec import Doc2Vec, TaggedDocument
tagged_docs = [TaggedDocument(words=doc, tags=[i]) for i, doc in enumerate(corpus_tokenized)]
```

```python
tagged_docs[0]
```

```python
len(tagged_docs)
```

```python
#TRAINING THE MODEL OVER OUR DATA
model = Doc2Vec(tagged_docs, vector_size=300, window=5, min_count=5, workers=4, epochs=20)
```

```python
#model dump:
MS_doc2vec_model = open("MS_doc2vec_model.pickle", "wb")
pickle.dump(model, MS_doc2vec_model)
MS_doc2vec_model.close()
```

```python
#NUMERICAL RRPRESENTATION OF TEXT
vector = model.infer_vector(corpus_tokenized[0])
```

```python
len(corpus_tokenized)
```

```python
import pandas as pd

# Create an empty DataFrame with the desired column names
df_data = pd.DataFrame(columns=['index', 'vector'])

# Iterate over the corpus_tokenized list and generate a vector for each element
for i, speech in enumerate(corpus_tokenized):
    vector = model.infer_vector(speech)
    # Store the index number and vector in a dictionary
    data = {'index': i, 'vector': vector}
    # Add the dictionary as a new row in the DataFrame
    df_data.loc[len(df_data)] = data
```

```python
# Extract the vectors as a list from the 'vector' column
vectors = df_data['vector'].tolist()
num_dimensions = len(vectors[0])
num_dimensions
```

```python
# Create column names with unique suffixes
column_names = ['dim_{}'.format(i) for i in range(1, num_dimensions + 1)]

# Create a list of dictionaries representing each row in the new DataFrame
new_data = [{column_names[i]: vector[i] for i in range(num_dimensions)} for vector in vectors]

# Create the new DataFrame
df_new = pd.DataFrame(new_data)
df_new
```

```python
MS_vector = open("MS_vector.pickle", "wb")
pickle.dump(df_new,MS_vector)
MS_vector.close()
```
