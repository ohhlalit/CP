We extract the URLs of the speeches given by Dr. Manmohan Singh from the website. It uses the BeautifulSoup library to parse the HTML code of the webpage, selects all the speech URLs from the HTML code using a CSS  selector, and then stores them in a list named 't'. The list 't' contains URLs of speeches given by Dr.Manmohan Singh.

```python
#LOOP FOR STORING THE URLs
doc = soupify(driver.page_source)
speeches = doc.select(".speechPan > ul > li > a")               
if len(speeches) == 0:
    pass
t = []                                                 # initialize the list 't'
for k in range(len(speeches)):
    target_speech = speeches[k]
    speech_url = target_speech.attrs['href']           #get the value of the 'href' attribute
    base_url = "https://archivepmo.nic.in/drmanmohansingh/"
    target_url = base_url + speech_url
    t.append(target_url)
print(t)  # or do whatever you want with the list 't'
```
